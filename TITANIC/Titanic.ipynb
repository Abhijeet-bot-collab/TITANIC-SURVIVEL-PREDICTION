EXPLORE AND PREPROCESS DATA

  - SUBTASK:-
      - Handle missing values in the 'Age', 'Fare', and 'Embarked' columns of both the train 
        and test datasets using appropriate imputation strategies.

 - Reasoning: Impute the missing values in 'Age', 'Fare', and 'Embarked' columns as per the 
   instructions.       

```python
train_data['Age'].fillna(train_data['Age'].mean(),inplace=True)
test_data['Age'].fillna(test_data['Age'].mean(), inplace=True)
test_data['Fare'].fillna(test_data['Fare'].mean(), inplace=True)
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
```

Reasoning: Verify that the missing values in the specified columns have been imputed correctly.

```python
print("\nMissing Values after impputation Tain Data:")
display(train_data[['Age', 'Embarked']].isnull().sum()) 

print("\nMissing Values after imputation Test Data:")
display(test_data[['Age', 'Fare']].isnull().sum())
```
