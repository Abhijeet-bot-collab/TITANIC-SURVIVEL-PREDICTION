````python
# Handle missing values
# Impute missing 'Age' values with the median age
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].median())
test_data['Age'] = test_data['Age'].fillna(test_data['Age'].median())

# Impute missing 'Fare' value with the median fare
test_data['Fare'] = test_data['Fare'].fillna(test_data['Fare'].median())

# Impute missing 'Embarked' values with the most frequent value
train_data['Embarked'] = train_data['Embarked'].fillna(train_data['Embarked'].mode()[0])

print("\nTrain data missing values after imputation:")
display(train_data.isnull().sum())

print("\nTest data missing values after imputation:")
display(test_data.isnull().sum())
````

FEATURE ENGINEERING 

````python
# CREATE 'TITLE' FEATURES FOR NAME 
train_data['Title'] = train_data['Name'].str.extract(r'([A-Za-z]+)\.', expand=False)
test_data['Title'] = test_data['Name'].str.extract(r'([A-Za-z]+)\.', expand=False)

# Standardize titles
train_data['Title'] = train_data['Title'].replace(['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
train_data['Title'] = train_data['Title'].replace('Mlle', 'Miss')
train_data['Title'] = train_data['Title'].replace('Ms', 'Miss')
train_data['Title'] = train_data['Title'].replace('Mme', 'Mrs')

test_data['Title'] = test_data['Title'].replace(['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
test_data['Title'] = test_data['Title'].replace('Mlle', 'Miss')
test_data['Title'] = test_data['Title'].replace('Ms', 'Miss')
test_data['Title'] = test_data['Title'].replace('Mme', 'Mrs')



print("\nTrain data Title counts:")
display(train_data['Title'].value_counts())

print("\nTest data Title counts:")
display(test_data['Title'].value_counts())

# Create 'family_size' feature
train_data['family_size'] = train_data['SibSp'] + train_data['Parch'] + 1
test_data['family_size'] = test_data['SibSp'] + test_data['Parch'] + 1

print("\nTrain data family_size counts:")
display(train_data['family_size'].value_counts())

print("\nTest data family_size counts:")
display(test_data['family_size'].value_counts())

# Create 'AgeBand' feature
train_data['AgeBand'] = pd.cut(train_data['Age'], 5)
test_data['AgeBand'] = pd.cut(test_data['Age'], 5)

print("\nTrain data AgeBand counts:")
display(train_data['AgeBand'].value_counts())

print("\nTest data AgeBand counts:")
display(test_data['AgeBand'].value_counts())

# Create 'FareBand' feature
train_data['FareBand'] = pd.qcut(train_data['Fare'], 4)
test_data['FareBand'] = pd.qcut(test_data['Fare'], 4)

print("\nTrain data FareBand counts:")
display(train_data['FareBand'].value_counts())

print("\nTest data FareBand counts:")
display(test_data['FareBand'].value_counts())

# Create 'Deck' feature from 'Cabin'
train_data['Deck'] = train_data['Cabin'].str.extract(r'([A-Za-z])', expand=False)
test_data['Deck'] = test_data['Cabin'].str.extract(r'([A-Za-z])', expand=False)

print("\nTrain data Deck counts:")
display(train_data['Deck'].value_counts())

print("\nTest data Deck counts:")
display(test_data['Deck'].value_counts())
````

DATA PREPROCESSING :- 
SUBTASK:
 - Apply one-hot encoding or other suitable encoders to categorical features, drop unnecessary columns,
    and set up a preprocessing pipeline using ColumnTransformer.

Reasoning: Identify categorical features, define columns to drop, create a ColumnTransformer for one-hot encoding, and apply it to both train and test data.      

````python
# Identify categorical features for encoding
categorical_features = ['Sex', 'Embarked', 'Pclass', 'Title', 'AgeBand', 'FareBand', 'Deck']

# Identify columns to drop
columns_to_drop = ['PassengerId', 'Name', 'Ticket', 'Cabin', 'SibSp', 'Parch', 'Age', 'Fare']

# Import necessary classes
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Create a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='drop' # Drop the remaining columns not specified in transformers
)

# Apply the ColumnTransformer to the training data
X_train_processed = preprocessor.fit_transform(train_data.drop('Survived', axis=1))

# Apply the same ColumnTransformer to the test data
X_test_processed = preprocessor.transform(test_data)

# Display the shapes of the preprocessed data
print("Shape of preprocessed training data:", X_train_processed.shape)
print("Shape of preprocessed test data:", X_test_processed.shape)
````

MODEL SELECTION AND TRAINING
 - SUBTASK :- 
     Establish a baseline model and then train at least three different models (e.g., Logistic Regression, RandomForest, XGBoost) using stratified 5-fold cross-validation.
 - Reasoning: 
     Define the target variable, import necessary libraries for model training and cross-validation, initialize models, and then iterate through the models to perform stratified k-fold cross-validation and store the results     

````python
# Define the target variable
y_train = train_data['Survived']

# Import necessary model classes and cross-validation tool
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score

# Initialize StratifiedKFold for cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Create a list of models to evaluate
models = [
    ('Logistic Regression', LogisticRegression(random_state=42, solver='liblinear')),
    ('Random Forest', RandomForestClassifier(random_state=42)),
    ('XGBoost', XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'))
]

# Store cross-validation results
cv_results = []

# Iterate through the models and perform cross-validation
for name, model in models:
    print(f"Training {name}...")
    cv_scores = cross_val_score(model, X_train_processed, y_train, cv=skf, scoring='accuracy')
    mean_accuracy = cv_scores.mean()
    cv_results.append((name, mean_accuracy))
    print(f"Mean accuracy for {name}: {mean_accuracy:.4f}\n")

# Identify the baseline best model
baseline_best_model_name, baseline_best_model_accuracy = max(cv_results, key=lambda item: item[1])

# Print the baseline best model
print(f"Baseline best model: {baseline_best_model_name} with a mean accuracy of {baseline_best_model_accuracy:.4f}")
````

MODEL EVALUATION 
SUBTASK :
  - Evaluate the performance of each model using metrics such as accuracy, precision, recall, F1-score, 
    and AUC.

REASONING :   Import the necessary metrics for model evaluation and initialize StratifiedKFold.  

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import StratifiedKFold

# Initialize StartfieldKFlod for cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
````

 Reasoning: Iterate through each model, perform cross-validation, calculate metrics for each fold, and store them.

````python
# Create dictionaries to store metrics for each model
model_accuracies = {}
model_precisions = {}
model_reasons = {}
model_f1s = {}
model_aucs = {}

# Define the target variable
y_train = train_data['Survived']

# Import necessary model classes and cross-validation tool
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score

# Initialize StratifiedKFold for cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Create a list of models to evaluate
models = [
    ('Logistic Regression', LogisticRegression(random_state=42, solver='liblinear')),
    ('Random Forest', RandomForestClassifier(random_state=42)),
    ('XGBoost', XGBClassifier(random_state=42, eval_metric='logloss'))
]

# Iterate through the models and perform cross-validation
for name, model in models:
    print(f"Evaluating {name}...")
    fold_accuracies = []
    fold_precisions = []
    fold_reasons = []
    fold_f1s = []
    fold_aucs = []

    for fold, (train_index, val_index) in enumerate(skf.split(X_train_processed, y_train)):
        X_train_fold, X_val_fold = X_train_processed[train_index], X_train_processed[val_index]
        y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]

        # Train the model
        model.fit(X_train_fold, y_train_fold)

        # Make predictions
        y_pred = model.predict(X_val_fold)

        # Calculate metrics
        accuracy = accuracy_score(y_val_fold, y_pred)
        precision = precision_score(y_val_fold, y_pred)
        recall = recall_score(y_val_fold, y_pred)
        f1 = f1_score(y_val_fold, y_pred)

        # Calculate AUC if the model supports predict_proba
        auc = np.nan # Initialize with NaN
        if hasattr(model, "predict_proba"):
            y_pred_proba = model.predict_proba(X_val_fold)[:, 1]
            auc = roc_auc_score(y_val_fold, y_pred_proba)

        fold_accuracies.append(accuracy)
        fold_precisions.append(precision)
        fold_reasons.append(recall)
        fold_f1s.append(f1)
        fold_aucs.append(auc)

    # Store mean metrics for the model
    model_accuracies[name] = np.mean(fold_accuracies)
    model_precisions[name] = np.mean(fold_precisions)
    model_reasons[name] = np.mean(fold_reasons)
    model_f1s[name] = np.mean(fold_f1s)
    model_aucs[name] = np.nanmean(fold_aucs) # Use nanmean to handle models without predict_proba

    print(f"Mean metrics for {name}:")
    print(f"  Accuracy: {model_accuracies[name]:.4f}")
    print(f"  Precision: {model_precisions[name]:.4f}")
    print(f"  Recall: {model_reasons[name]:.4f}")
    print(f"  F1-score: {model_f1s[name]:.4f}")
    if not np.isnan(model_aucs[name]):
        print(f"  AUC: {model_aucs[name]:.4f}")
    else:
        print("  AUC: Not applicable")
    print("-" * 20)
````

HYPERPARAMETER TUNING
SUBTASK:
  - Perform hyperparameter tuning using Grid Search or Random Search on the best performing models.

Reasoning: Identify the best performing model and define a hyperparameter grid for it, then use GridSearchCV to find the best hyperparameters.

````python
from sklearn.model_selection import GridSearchCV

# Identify the best performing model based on mean accuracy from the previous step
# Based on the printed output from the previous cell:
# Mean accuracy for Logistic Regression: 0.8092
# Mean accuracy for Random Forest: 0.8182
# Mean accuracy for XGBoost: 0.8204
# XGBoost had the highest mean accuracy.

best_model = XGBClassifier(random_state=42, eval_metric='logloss')

# Define a hyperparameter grid for XGBoost
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0],
    'gamma': [0, 0.1, 0.2]
}

# Instantiate GridSearchCV
grid_search = GridSearchCV(estimator=best_model, param_grid=param_grid, cv=skf, scoring='accuracy', n_jobs=-1)

# Fit GridSearchCV to the preprocessed training data
grid_search.fit(X_train_processed, y_train)

# Print the best hyperparameters and best score
print("\nBest hyperparameters found by Grid Search:")
print(grid_search.best_params_)
print("\nBest cross-validation accuracy:")
print(grid_search.best_score_)
````

FINAL MODEL EVALUATION 
SUBTASK : 
 - Evaluate the best model after tuning using a confusion matrix and ROC curve.

Reasoning: Retrieve the best estimator, make predictions, generate and plot the confusion matrix, calculate predicted probabilities, calculate ROC curve components and AUC, and plot the ROC curve to evaluate the best model after tuning.

````python
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score

# 1. Retrieve the best estimator
best_model_tuned = grid_search.best_estimator_

# 2. Make predictions on the entire preprocessed training data
y_pred_tuned = best_model_tuned.predict(X_train_processed)

# 3. Generate and plot a confusion matrix
cm = confusion_matrix(y_train, y_pred_tuned)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Survived', 'Survived'], yticklabels=['Not Survived', 'Survived'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix of Tuned XGBoost Model')
plt.show()

# 4. Calculate the predicted probabilities for the positive class
y_pred_proba_tuned = best_model_tuned.predict_proba(X_train_processed)[:, 1]

# 5. Calculate the False Positive Rate (FPR), True Positive Rate (TPR), and thresholds for the ROC curve
fpr, tpr, thresholds = roc_curve(y_train, y_pred_proba_tuned)

# 6. Calculate the Area Under the ROC Curve (AUC)
auc_score = roc_auc_score(y_train, y_pred_proba_tuned)

# 7. Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {auc_score:.4f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Random Classifier')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve of Tuned XGBoost Model')
plt.legend()
plt.grid(True)
plt.show()
````
